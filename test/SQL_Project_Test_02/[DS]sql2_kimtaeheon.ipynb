문제 3.

다음의 함수와 그 함수를 테스트하는 코드를 작성하세요. (PDF 파일 참고)

- 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수 (테스트 입력 : ‘1,000’)
- 주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID 를 반환하는 함수 (테스트 입력 : ‘SK에너지’) - 주소를 입력받아 구 이름을 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
- 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)

제출 3.
- 각 함수 및 테스트 코드와 실행결과

```python
import googlemaps


# 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수
def str2int(str) :
    if str != '' :
        return int(''.join(str.split(',')))

# 주유소 브랜드를 입력하면 GAS_BRAND 데이터를 참고하여 ID를 반환하는 함수
def search_id(gas_brand) :
    cursor.execute("select * from GAS_BRAND")
    result = cursor.fetchall()

    for result_iterator in result :
        if gas_brand in result_iterator[1] :
            id = result_iterator[0]
        else :
            if gas_brand == 'HD현대오일뱅크' :
                id = 2
            elif gas_brand == '알뜰(ex)' :
                id = 5

    return id


# 주소를 입력받아 구 이름을 반환하는 함수
def search_gu(address) :
    split_address = address.split()
    gu = split_address[1]

    return gu

# 주소를 입력받아 위도, 경도를 반환하는 함수
def search_latlng(address) :
    gmaps_key = "AIzaSyAApXA75cS4g5TY6mZ55LAltw4o4rMFuvE"
    gmaps = googlemaps.Client(key=gmaps_key)
    map_info = gmaps.geocode(address, language='ko')
    lat = map_info[0].get('geometry')['location']['lat']
    lng = map_info[0].get('geometry')['location']['lng']
    
    return lat, lng
```

```python
# 함수 결과 출력

# 화폐단위 문자형을 입력받아 숫자형으로 반환하는 함수
result1 = str2int('1,000')
result2 = search_id('SK에너지')
result3 = search_gu('서울시 강남구 헌릉로 730')
result4 = search_latlng('서울시 강남구 헌릉로 730')

print(result1)
print(result2)
print(result3)
print(result4)
```

문제 4.

Python 코드에서 주유소 페이지에서 데이터를 가져올때, GAS_STATION 테이블에 바로 입력하도록 수정하세요. (앞서 생 성한 함수활용)
- 주의. city 는 ‘서울’ 로 고정, 부가정보 데이터 타입
- 입력된 데이터의 총 갯수를 쿼리하여 결과를 확인합니다. - 입력된 데이터 상위 10개를 쿼리하여 결과를 확인합니다.


제출 4.
- 주유소 데이터 관련 코드 (ipynb)



```python
# 제출 4
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from tqdm import tqdm
from bs4 import BeautifulSoup as bs

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
```

```python
# 시 선택 박스를 서울로 맞춰주는 작업
sido_list_raw = driver.find_element(By.ID, "SIDO_NM0")
sido_list = sido_list_raw.find_elements(By.TAG_NAME, "option")
sido_names = [option.get_attribute("value") for option in sido_list]
sido_list_raw.send_keys(sido_names[1])
```

```python
# 서울시의 각 구 이름
gu_list_raw = driver.find_element(By.ID, "SIGUNGU_NM0")
gu_list = gu_list_raw.find_elements(By.TAG_NAME, "option")
gu_names = [option.get_attribute("value") for option in gu_list]
gu_names = gu_names[1:]
```

```python
# 각 구를 입력하여 서울시 모든 구의 주유소 정보 가져오기
# 가져온 데이터는 바로 GAS_STATION 테이블에 insert
sql = "insert into GAS_STATION (brand, name, city, gu, address, gasoline, diesel, self, car_wach, charging_station, car_maintenance, convenience_store, 24_hours, lat, lng) values (%s, %s, '서울', %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"

for gu in tqdm(gu_names) :
    # 시/군/구 태그 선택
    gu_element = driver.find_element(By.ID, 'SIGUNGU_NM0')
    gu_element.send_keys(gu)
    time.sleep(0.5)

    # 시/군/구 선택 시 결과 화면 html 파싱
    html = driver.page_source
    soup = bs(html, "html.parser")

    # 검색 결과 수만큼 반복
    station_cnt = int(soup.select_one('#totCnt').text)
    for i in range(1, station_cnt+1) :
        each_station = driver.find_element(By.CSS_SELECTOR, f"#body1 > tr:nth-child({i}) > td.rlist > a").click()
        time.sleep(0.5)

        # 각 주유소 정보 화면에 대한 html 파싱
        html = driver.page_source
        soup = bs(html, "html.parser")

        # 브랜드
        station_brand = search_id(soup.select_one("#poll_div_nm").text)

        # 주유소명
        station_name = soup.select_one("#os_nm").text

        # 주소
        station_address = soup.select_one("#rd_addr").text

        # 구
        station_gu = search_gu(station_address)

        # 휘발유 가격
        gasoline_price = str2int(soup.select_one("#b027_p").text)

        # 경유 가격
        diesel_price = str2int(soup.select_one("#d047_p").text)

        # 셀프 여부
        self_yn = True if soup.find(class_='header').find(id="SPAN_SELF_VLT_YN_ID").find(id="self_icon") else False

        # 세차장 여부
        wash_yn = True if "off" not in soup.select_one("#cwsh_yn")['src'] else False

        # 충전소 여부
        lpg_yn = True if "off" not in soup.select_one("#lpg_yn")['src'] else False

        # 경정비 여부
        maint_yn = True if "off" not in soup.select_one("#maint_yn")['src'] else False

        # 편의점 여부
        cvs_yn = True if "off" not in soup.select_one("#cvs_yn")['src'] else False

        # 24시간 여부
        sel24_yn = True if "off" not in soup.select_one("#sel24_yn")['src'] else False

        # 위도
        lat = search_latlng(station_address)[0]

        # 경도
        lng = search_latlng(station_address)[1]

        cursor.execute(sql, (station_brand, station_name, station_gu, station_address,
                            gasoline_price, diesel_price, self_yn, wash_yn, lpg_yn, maint_yn, cvs_yn,
                            sel24_yn, lat, lng))
        
        remote.commit()

        time.sleep(0.5)
```

문제 6.

위도, 경도 정보를 이용하여 미왕빌딩에서 1킬로 이내에 위치한 주유소 정보를 검색하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 미왕빌딩으로 부터의 거리 (km)

제출 6.
- 쿼리, 결과

```python
cursor.execute("set @location = point(127.02915553846, 37.495435686811)")   # 미왕빌딩 경도, 위도
# ST_Distance_Sphere 결과 m단위이기 때문에 1000을 나눠 km로 변경
cursor.execute("select s.id, b.name 'brand', s.name, s.address, ST_Distance_Sphere(@location, point(lng, lat))/1000 distance from GAS_BRAND b, GAS_STATION s where b.id=s.brand having distance <= 1")
result_distance_under_one = cursor.fetchall()
for datas in result_distance_under_one:
    print(datas)
```

문제 7.

위도, 경도 정보를 이용하여 미왕빌딩에서 셀프주유가 가능하고 24시간이면서 편의점이 있는 가장 가까운 주유소 10개를 휘 발유 가격이 가장 저렴한 순으로 정렬하여 조회하세요.

- 주유소 아이디, 주유소 브랜드명, 주유소 매장명, 주소, 휘발유 가격, 부가정보 (셀프, 24시간, 편의점 여부), 미왕빌딩으로 부터의 거리 (km)


제출 7.
- 쿼리, 결과

```python
cursor.execute("set @location = point(127.02915553846, 37.495435686811)")
cursor.execute("select * from (select s.id, b.name 'brand', s.name, s.address, s.gasoline, s.self, s.24_hours, s.convenience_store, ST_Distance_Sphere(@location, point(lng, lat))/1000 distance from GAS_STATION s, GAS_BRAND b where b.id=s.brand and s.self=1 and s.24_hours=1 and s.convenience_store=1 order by distance limit 10) new_table order by new_table.gasoline")
result_top_10 = cursor.fetchall()
for datas in result_top_10 :
    print(datas)
```

문제 8.

구별로 주유소 브랜드 별 휘발유 평균가격을 조회하여 저렴한 순으로 출력하세요.

- 구 이름, 주유소 브랜드 이름, 휘발유 평균 가격

제출 8.
- 쿼리, 결과

```python
cursor.execute("select s.gu, b.name 'brand', avg(gasoline) from GAS_STATION s, GAS_BRAND b where b.id=s.brand group by gu, brand order by avg(gasoline)")
result_avg_gasoline = cursor.fetchall()
for datas in result_avg_gasoline :
    print(datas)
```
