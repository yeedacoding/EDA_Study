````python
# 2-1
import warnings
warnings.filterwarnings('ignore')

# 조건1 : 종로구
df_jongro = df_target[df_target['자치구']=='종로구']

# 기준이 되는 행 추출
std_jongro = df_jongro[12:]
jongro_pop = std_jongro[std_jongro['구분']=='인구']
jongro_density = std_jongro[std_jongro['구분']=='인구밀도']
jongro_speed = std_jongro[std_jongro['구분']=='차량통행속도']

tmp_df = df_jongro.drop(std_jongro.index)

# 조건3
acc_per_pop = deepcopy(tmp_df)
acc_per_pop['구분'] = '인구 만명당 ' + acc_per_pop['구분'].astype(str)

for idx, row in acc_per_pop.iterrows() :
    for column in acc_per_pop.columns[3:] :
        acc_per_pop.at[idx, column] = round((row[column] / jongro_pop[column]) * 10000, 2)

# 조건4
acc_per_density = deepcopy(tmp_df)
acc_per_density['구분'] = '인구밀도당 ' + acc_per_density['구분'].astype(str)

for idx, row in acc_per_density.iterrows() :
    for column in acc_per_density.columns[3:] :
        acc_per_density.at[idx, column] = round((row[column] / jongro_density[column]), 2)

# 조건5
acc_per_speed = deepcopy(tmp_df)
acc_per_speed['구분'] = '차량통행속도당 ' + acc_per_speed['구분'].astype(str)

for idx, row in acc_per_speed.iterrows() :
    for column in acc_per_speed.columns[3:] :
        acc_per_speed.at[idx, column] = round((row[column] / jongro_speed[column]), 2)

# 조건6
tmp_df = pd.concat([acc_per_pop, acc_per_density, acc_per_speed])

# 조건7
tmp_df['사고유형'] = pd.Categorical(tmp_df['사고유형'], categories=type_list, ordered=True)
tmp_df['구분'] = pd.Categorical(tmp_df['구분'], categories=cat_list, ordered=True)

tmp_df = tmp_df.sort_values(['사고유형', '구분'])

df_result = tmp_df.reset_index(drop=True)

check_02_01(df_result)
````

````python
gu_list = ['서울시', '종로구', '중구', '용산구', '성동구', '광진구', '동대문구', '중랑구', '성북구', '강북구', '도봉구', '노원구', '은평구', '서대문구', '마포구', '양천구', '강서구', '구로구', '금천구', '영등포구', '동작구', '관악구', '서초구', '강남구', '송파구', '강동구']
cat_list = ['인구 만명당 발생건수', '인구 만명당 사망자수', '인구 만명당 부상자수', '인구밀도당 발생건수', '인구밀도당 사망자수', '인구밀도당 부상자수',  '차량통행속도당 발생건수', '차량통행속도당 사망자수',  '차량통행속도당 부상자수']
type_list = ['합계', '차대사람', '차대차', '차량단독', '건널목']
````

````python
# 2-2
import warnings
warnings.filterwarnings('ignore')

# 서울 전체
# 2-1 코드 함수화

# 기준이 되는 행 추출
def preprocessing_std(gu) :
    tmp_city = df_target[df_target['자치구'] == gu]

    std_city = tmp_city[tmp_city['사고유형'] == '-']
    city_pop = std_city[std_city['구분']=='인구']
    city_density = std_city[std_city['구분']=='인구밀도']
    city_speed = std_city[std_city['구분']=='차량통행속도']

    tmp_df = tmp_city.drop(std_city.index)

    return city_pop, city_density, city_speed, tmp_df

# 조건3
# 구별 인구 만명당 교통사고율
def acc_pop_df(tmp_df, city_pop) :
    acc_per_pop = deepcopy(tmp_df)
    acc_per_pop['구분'] = '인구 만명당 ' + acc_per_pop['구분'].astype(str)

    for idx, row in acc_per_pop.iterrows() :
        for column in acc_per_pop.columns[3:] :
            acc_per_pop.at[idx, column] = ((row[column] / city_pop[column]) * 10000).round(2)
    
    return acc_per_pop

# 조건4
# 구별 인구밀도당 교통사고율
def acc_density_df(tmp_df, city_density) :
    acc_per_density = deepcopy(tmp_df)
    acc_per_density['구분'] = '인구밀도당 ' + acc_per_density['구분'].astype(str)

    for idx, row in acc_per_density.iterrows() :
        for column in acc_per_density.columns[3:] :
            acc_per_density.at[idx, column] = (row[column] / city_density[column]).round(2)

    return acc_per_density

# 조건5
# 구별 차량통행속도당 교통사고율
def acc_speed_df(tmp_df, city_speed) :
    acc_per_speed = deepcopy(tmp_df)
    acc_per_speed['구분'] = '차량통행속도당 ' + acc_per_speed['구분'].astype(str)

    for idx, row in acc_per_speed.iterrows() :
        for column in acc_per_speed.columns[3:] :
            acc_per_speed.at[idx, column] = (row[column] / city_speed[column]).round(2)
    
    return acc_per_speed

# 조건6
# 구별 조건3, 4, 5 데이터프레임 concat
def concat_per_gu(acc_per_pop, acc_per_density, acc_per_speed) :
    concat_df = pd.concat([acc_per_pop, acc_per_density, acc_per_speed])

    return concat_df

# 조건7
def concat_all_gu(li) :
    concat_df = pd.concat(li)

    concat_df['자치구'] = pd.Categorical(concat_df['자치구'], categories=gu_list, ordered=True)
    concat_df['사고유형'] = pd.Categorical(concat_df['사고유형'], categories=type_list, ordered=True)
    concat_df['구분'] = pd.Categorical(concat_df['구분'], categories=cat_list, ordered=True)

    concat_df = concat_df.sort_values(['자치구', '사고유형', '구분'])

    df_result = concat_df.reset_index(drop=True)

    return df_result

# 서울 전체 구 코드 실행
df_list = []

for gu in gu_list :
    city_pop, city_density, city_speed, tmp_df = preprocessing_std(gu)
    acc_per_pop = acc_pop_df(tmp_df, city_pop)
    acc_per_density = acc_density_df(tmp_df, city_density)
    acc_per_speed = acc_speed_df(tmp_df, city_speed)

    concat_gu = concat_per_gu(acc_per_pop, acc_per_density, acc_per_speed)
    
    df_list.append(concat_gu)

df_result = concat_all_gu(df_list)

check_02_02(df_result)
````
