```python
df_target_temp = df_target.copy()

def calc_time(data, weekday) :
    data[f'{weekday}관람가능시간'] = 0

    for idx, row in data.iterrows() :
        s_hour = int(row[f'{weekday}관람시작시각'].split(':')[0])
        s_min = int(row[f'{weekday}관람시작시각'].split(':')[1])
        e_hour = int(row[f'{weekday}관람종료시각'].split(':')[0])
        e_min = int(row[f'{weekday}관람종료시각'].split(':')[1])

        min = round(e_min / 60, 2) - round(s_min / 60, 2)
        if min < 0 :
            hour = e_hour - s_hour - 1
            min = abs(min)
            total_time = float(hour + min)
        else :
            hour = e_hour - s_hour
            total_time = float(hour + min)

        if total_time > 23 :
            total_time = 24

        data.loc[idx, f'{weekday}관람가능시간'] = total_time
    
    return data

df_target_temp = calc_time(df_target_temp, '평일')
df_target_temp = calc_time(df_target_temp, '공휴일')

df_target_temp.head()
# df_target_temp['평일관람종료시각'] - df_target_temp['평일관람시작시각']
```
