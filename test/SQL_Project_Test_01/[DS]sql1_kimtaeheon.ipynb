Python 코드로 스타벅스 페이지에서 데이터를 가져올때, COFFEE_STORE 테이블에 바로 입력하도록 수정하세요.

- 데이터 세트: 매장 이름, 매장이 위치한 구 이름, 매장 주소, 위도, 경도
- 필요한 데이터를 한세트씩 가져와서 COFFEE_STORE 테이블에 각각INSERT 하도록 합니다.
- 입력된 데이터의 총 갯수를 쿼리하여 결과를 확인합니다.
- 입력된 데이터 상위 10개를 쿼리하여 결과를 확인합니다.

```python
import time
from bs4 import BeautifulSoup as bs
from tqdm import tqdm
```

```python
# '지역 검색' 태그 찾아서 클릭
search_region_tag = driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > header.loca_search > h3 > a")
search_region_tag.click()
time.sleep(0.5)

# '서울' 태그 찾아서 클릭
seoul_tag = driver.find_element(By.CSS_SELECTOR, "#container > div > form > fieldset > div > section > article.find_store_cont > article > article:nth-child(4) > div.loca_step1 > div.loca_step1_cont > ul > li:nth-child(1) > a")
seoul_tag.click()
time.sleep(0.5)

# '전체' 태그 찾아서 클릭
gu_all_tag = driver.find_element(By.CSS_SELECTOR, "#mCSB_2_container > ul > li:nth-child(1) > a")
gu_all_tag.click()
time.sleep(0.5)

# '전체' 태그 선택 결과를 BeautifulSoup를 활용하여 html로 파싱
html = driver.page_source
soup = bs(html, 'html.parser')

# '서울 - 전체' 스타벅스 매장 정보 호출
datas = soup.select("#mCSB_3_container > ul > li")

sql = "insert into COFFEE_STORE (brand, name, gu_name, address, lat, lng) values (1, %s, %s, %s, %s, %s)"

# 각 매장별 매장 이름, 매장이 위치한 구 이름, 매장 주소, 위도, 경도 가져오기
for data in tqdm(datas) :
    name = data.select_one('strong').get_text().strip()
    address = data.select_one('p').get_text()[:-9]
    gu = address.split()[1]
    lat = data['data-lat']
    lng = data['data-long']

    cursor.execute(sql, (name, gu, address, lat, lng))
    remote.commit()

    time.sleep(0.5)

driver.quit()
```

문제 6.

Python 코드로 이디야 페이지에서 데이터를 가져올때, COFFEE_STORE 테이블에 바로 입력하도록 수정하세요.

- 데이터 세트 : 매장 이름, 매장이 위치한 구 이름, 매장 주소, 위도, 경도
- 이디야 페이지에서 검색에 사용할 구 이름은 COFFEE_STORE 에서 중복을 제거하는 쿼리를 사용하여 가져와서 {‘서울 ‘ + 구이름} 형식으로 변환하여 사용하도록 합니다.
- 필요한 데이터를 한 세트씩 가져와서 COFFEE_STORE 테이블에 각각 INSERT 하도록 합니다.
- (주의) COFFEE_STORE 테이블에 입력할 구 이름은 {‘서울 ‘} 이 제거된 구 이름입니다.
- 입력된 데이터의 총 갯수를 쿼리하여 결과를 확인합니다.
- 입력된 데이터 상위 10개를 쿼리하여 결과를 확인합니다.

```python
# 이디야 커피 url 접근
driver = webdriver.Chrome()
driver.get("https://www.ediya.com/contents/find_store.html#c")
time.sleep(10)

# '주소' 태그 찾아서 클릭
search_address_tag = driver.find_element(By.CSS_SELECTOR, '#contentWrap > div.contents > div > div.store_search_pop > ul > li:nth-child(2) > a')
search_address_tag.click()
```

```python
# 이디야 페이지에서 검색에 사용할 구 이름은 COFFEE_STORE 에서 중복을 제거하는 쿼리를 사용하여 가져와서 {‘서울 ‘ + 구이름} 형식으로 변환하여 사용하도록 합니다.
cursor.execute("select distinct concat ('서울 ', gu_name) from COFFEE_STORE")

address_list = []
address_result = cursor.fetchall()

for address in address_result :
    address_list.append(address[0])
```

```python
# 위도, 경도값을 가져오기 위한 google map api 호출
import googlemaps
import numpy as np

gmaps_key = "<gmaps api key>"
gmaps = googlemaps.Client(gmaps_key)
```

```python
# 데이터 세트 : 매장 이름, 매장이 위치한 구 이름, 매장 주소, 위도, 경도
# 필요한 데이터를 한 세트씩 가져와서 COFFEE_STORE 테이블에 각각 INSERT 하도록 합니다.
# (주의) COFFEE_STORE 테이블에 입력할 구 이름은 {‘서울 ‘} 이 제거된 구 이름입니다.
sql = 'insert into COFFEE_STORE (brand, name, gu_name, address, lat, lng) values (2, %s, %s, %s, %s, %s)'

for address in tqdm(address_list) :
    # 주소 입력 칸 태그를 찾고 새로운 주소 입력 후 돋보기 버튼 클릭
    input_address = driver.find_element(By.CSS_SELECTOR, "#keyword")
    input_address.send_keys(address)
    time.sleep(0.5)

    submit_btn = driver.find_element(By.CSS_SELECTOR, '#keyword_div > form > button')
    submit_btn.click()
    input_address.clear()
    time.sleep(0.5)

    # 클릭 결과 화면 HTML 파싱
    html = driver.page_source
    soup = bs(html, 'html.parser')
    datas = soup.select("#placesList > li.item")
    
    for data in datas :
        cafe_info = data.select_one("a").get_text()
        name = cafe_info.split()[0]
        address = cafe_info[len(name):].strip()
        gu = address.split()[1]

        tmp = gmaps.geocode(address, language='ko')
        lat = tmp[0].get('geometry')['location']['lat'] if tmp else np.nan
        lng = tmp[0].get('geometry')['location']['lng'] if tmp else np.nan

        cursor.execute(sql, (name, gu, address, lat, lng))
        remote.commit()

        time.sleep(0.5)

driver.quit()
```

문제 8.

시각화 프로젝트를 위하여 다음의 규칙으로 쿼리하여 CSV 파일로 저장합니다. (Python 코드로 작업)

- 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다.

```python
# 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
# 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
import pandas as pd
cursor.execute("select starbucks.*, ediya.* from (select * from COFFEE_STORE where brand=1) as starbucks join (select * from COFFEE_STORE where brand=2) as ediya on starbucks.gu_name=ediya.gu_name")
datas = cursor.fetchall()

df = pd.DataFrame(datas)

df.columns = ['s_id', 's_brand', 's_name', 's_gu', 's_address', 's_lat', 's_lng',
              'e_id', 'e_brand', 'e_name', 'e_gu', 'e_address', 'e_lat', 'e_lng']

df = df.sort_values(by=['s_id', 'e_id'])
df = df.reset_index(drop=True)
df['s_brand'] = 'STARBUCKS'
df['e_brand'] = 'EDIYA'
```
