    - 참고
        - Coding 방식에 따라 **SettingWithCopyWarning** 이 발생할 수 있습니다.
        - 해당 경고(Warning)가 발생하더라도 정답 인정 여부와는 상관 없습니다.

<br></br>
- 완료 후 결과 DataFrame 변수(df_target)를 check_01_05 함수에 입력하여 채점하세요.

````python
# 1-5
change_columns = df_target.iloc[:, :17].columns
gu_indexes = df_target.index.get_level_values(0).unique()

for gu in gu_indexes : 
    for column in change_columns :
        if len(df_target.loc[gu].index.get_level_values(0).unique()) == 5:
            df_target[column].loc[(gu, '합계', '발생건수')] = df_target[column].loc[(gu, '차대사람', '발생건수')] + df_target[column].loc[(gu, '차대차', '발생건수')] + df_target[column].loc[(gu, '차량단독', '발생건수')] + df_target[column].loc[(gu, '건널목', '발생건수')]
        else :
            df_target[column].loc[(gu, '합계', '발생건수')] = df_target[column].loc[(gu, '차대사람', '발생건수')] + df_target[column].loc[(gu, '차대차', '발생건수')] + df_target[column].loc[(gu, '차량단독', '발생건수')]

check_01_05(df_target)
````
