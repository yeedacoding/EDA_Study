## folium.Marker(location, popup, tooltip)
- 지도에 마커 생성
- location 
- popup : 마커 클릭 시 레이블 생성(str), html 문법 사용 가능
- tooltip : 마커에 마우스 호버 시 레이블 생성, html 문법 사용 가능

```python
m = folium.Map(location=[37.532077, 126.979725],    # 용산구
               zoom_start = 15,
               tiles="Cartodb Positron")

# 용산
folium.Marker(
    location=(37.532077, 126.979725)
).add_to(m)

# 전쟁기념관
folium.Marker(
    location=(37.537488, 126.977278),
    popup='<b>전쟁기념관</b>'
).add_to(m)

# 이태원시장
folium.Marker(
    location=(37.535310, 126.989938),
    tooltip='<i>이태원 시장</i>'
).add_to(m)

# 국립중앙박물관
folium.Marker(
    location=(37.524276, 126.980344),
    popup='<a href="https://www.museum.go.kr/site/eng/content/hours_and_fees", target=_"blink">국립중앙박물관</a>'
).add_to(m)

m
```

## folium.Marker(location, icon)
1. `icon_color` : str, default 'white'
- The color of the drawing on the marker. You can use colors above, or an html color code.
2. `icon` : str, default 'info-sign'
- The name of the marker sign. See Font-Awesome website to choose yours. Warning : depending on the icon you choose you may need to adapt the `prefix` as well.
3. `angle` : int, default 0. 
- The icon will be rotated by this amount of degrees.
4. `prefix` : str, default 'glyphicon'. 
- The prefix states the source of the icon. 'fa' for font-awesome or. 'glyphicon' for bootstrap 3.
- fa : https://fontawesome.com/

```python
m = folium.Map(location=[37.532077, 126.979725],    # 용산구
               zoom_start = 15,
               tiles="Cartodb Positron")

# 용산
# icon = folium.Icon(color, icon)
folium.Marker(
    location=(37.532077, 126.979725),
    icon=folium.Icon(
        color='black',
        icon='info-sign'
    )
).add_to(m)

# 전쟁기념관
# icon = folium.Icon(color, icon_color, icon)
folium.Marker(
    location=(37.537488, 126.977278),
    popup='<b>전쟁기념관</b>',
    icon=folium.Icon(
        color='red',
        icon_color='pink',
        icon='cloud'
    )
).add_to(m)

# 이태원시장
# Icon custom
folium.Marker(
    location=(37.535310, 126.989938),
    tooltip='<i>이태원 시장</i>',
    icon=folium.Icon(
        color='blue',
        icon_color='yellow',
        icon='fire',
        angle=50,
        prefix='fa'
    )
).add_to(m)

# 국립중앙박물관
folium.Marker(
    location=(37.524276, 126.980344),
    popup='<a href="https://www.museum.go.kr/site/eng/content/hours_and_fees", target=_"blink">국립중앙박물관</a>'
).add_to(m)

m
```

## folium.ClickForMarker()
- 지도 위에 마우스 클릭 시 클릭한 위치에 마커 생성
- 생성된 마커는 popup기능을 가지고 있으며, 레이블은 해당 마커의 위도, 경도 값이다.
- popup 옵션을 지정하여 원하는 레이블로 수정 가능하다.

```python
m = folium.Map(location=[37.532077, 126.979725],    # 용산구
               zoom_start = 15,
               tiles="Cartodb Positron")

m.add_child(folium.ClickForMarker())
```

## folium.Circle()
- 기본 사용법은 folium.Marker와 같으며, 원의 반지름(radius) 옵션만 지정해주면 된다.
- radius : 원의 반지름 값
- color : 원 색상
- fill : 원 내부를 채울 건지에 대한 옵션
- fill_color : 원 내부 색상

## folium.CircleMarker()
- 기본 사용법은 folium.Circle()과 동일
- 지도의 확대 축소에 따라 원의 넓이가 달라짐

```python
m = folium.Map(location=[37.532077, 126.979725],    # 용산구
               zoom_start = 14,
               tiles="Cartodb Positron")

# 용산
folium.Circle(
    location=(37.532077, 126.979725),
    radius=100
).add_to(m)

# 전쟁기념관
folium.Circle(
    location=(37.537488, 126.977278),
    radius=100,
    color='green',
    fill=True,
).add_to(m)

# 이태원시장
folium.Circle(
    location=(37.535310, 126.989938),
    popup='이태원시장 circle',
    radius=100,
    color='red',
    fill=True,
    fill_color='orange'
).add_to(m)

# 국립중앙박물관
folium.CircleMarker(
    location=(37.524276, 126.980344),
    popup='국립중앙박물관',
    radius=100,
    color='yellow',
    fill=True
).add_to(m)

m
```

## folium.Choropleth()
- 특정 구역의 경계값을 갖는 geoJson을 통해 지도 위에 경계를 overlay 해주는 기능
- geo_data : 경계값을 갖는 파일, url 등등
- data : series나 dataframe
- columns : data 옵션에 전달된 데이터의 컬럼
- fill_color : 색상
- fill_opacity : 색상 투명도
- line_opacity : 라인 투명도

```python
import json

# 미국 2012년 10월 실업률 현황 데이터 
state_data = pd.read_csv('../data/data02/02. US_Unemployment_Oct2012.csv')

state_data.head()
```

```python
m = folium.Map(
    location=[43, -102],
    zoom_start=3
)

folium.Choropleth(
    geo_data='../data/data02/02. us-states.json',    # 경계선 좌표값이 담긴 데이터
    data=state_data,
    columns=['State', 'Unemployment'],
    key_on='feature.id',
    fill_color='BuPu',
    fill_opacity=1,
    line_opacity=1,
    legend_name='Unemployment rate(%)'
).add_to(m)

m
```
