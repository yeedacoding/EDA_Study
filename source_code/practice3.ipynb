```python
from selenium import webdriver

driver = webdriver.Chrome()
driver.get("https://pinkwink.kr/") 
```

# 화면 크기 설정

```python
# 최대 크기로 설정
driver.maximize_window()
```

```python
# 창 최소화
driver.minimize_window()
```

```python
# 특정 값으로 창 크기 설정
driver.set_window_size(800, 800)
```

# 새로고침, 뒤로 가기, 앞으로 가기

```python
# 새로고침
driver.refresh()
```

```python
# 뒤로 가기
driver.back()
```

```python
# 앞으로 가기
driver.forward()
```

# 클릭

```python
from selenium.webdriver.common.by import By

first_content = driver.find_element(By.CSS_SELECTOR, "#content > div.cover-masonry > div > ul > li:nth-child(2)")

first_content.click()
```

# 탭 동작

```python
# 새 탭 생성
driver.execute_script("window.open()")
```

```python
# 새 탭 네이버 생성
driver.execute_script("window.open('https://www.naver.com')")
```

```python
# 탭 이동
driver.switch_to.window(driver.window_handles[0])
```

```python
# 탭 닫기
driver.close()
```

```python
driver.quit()
```

# 화면 스크롤

```python
# 스크롤할 수 있는 html body의 높이값 반환 
driver.execute_script('return document.body.scrollHeight')
```

```python
# 가장 하단으로 이동
driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
```

```python
# 가장 상단으로 이동
driver.execute_script("window.scrollTo(0, 0)")
```

```python
# 특정 태그 위치로 스크롤 이동
from selenium.webdriver.common.by import By
from selenium.webdriver import ActionChains

some_tag = driver.find_element(By.CSS_SELECTOR, "#content > div.cover-list > div > ul > li:nth-child(1)")
action = ActionChains(driver)
action.move_to_element(some_tag).perform()
```

# 검색어 입력

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

driver.get("https://www.naver.com") 
```

```python
# 검색창에 검색어 입력
keyword = driver.find_element(By.CSS_SELECTOR, "#query")
keyword.send_keys("파이썬")
```

```python
# 검색 버튼 선택 후 클릭
search_btn = driver.find_element(By.CSS_SELECTOR, "#sform > fieldset > button")
search_btn.click()
```

# 검색어 지우기

```python
keyword = driver.find_element(By.CSS_SELECTOR, "#query")
keyword.clear()
keyword.send_keys("딥러닝")
```

```python
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()

driver.get("https://pinkwink.kr/")
```

```python
# 돋보기 버튼 선택
from selenium.webdriver import ActionChains

search_btn = driver.find_element(By.CSS_SELECTOR, '.search')
action = ActionChains(driver)
action.click(search_btn).perform()
```

```python
driver.find_element(By.CSS_SELECTOR, '#header > div.search.on > input[type=text]').send_keys("딥러닝")
```

```python
driver.find_element(By.CSS_SELECTOR, '#header > div.search.on > button').click()
```

# selenium과 BeautifulSoup 같이 쓰기
