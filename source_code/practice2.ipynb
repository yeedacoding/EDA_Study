```python
import requests
from bs4 import BeautifulSoup as bs

url = 'https://finance.naver.com/marketindex/'
res = requests.get(url)

res
```

```python
res.content
```

```python
soup = bs(res.content, "html.parser")

print(soup.prettify())
```

```python
# class : .
# id : #
exchangeList = soup.select("#exchangeList > li")

print(exchangeList)
print(len(exchangeList))
```

```python
title = exchangeList[0].select_one(".h_lst").get_text()
exchange = exchangeList[0].select_one(".value").get_text()
change = exchangeList[0].select_one(".change").get_text()
updown = exchangeList[0].select_one(".head_info.point_dn > .blind").get_text()

title, exchange, change, updown
```

```python
link = exchangeList[0].select_one("a").get('href')

link
```

```python
baseUrl = 'https://finance.naver.com/'
print(baseUrl + link)
```

```python
exchange_datas = []
baseUrl = 'https://finance.naver.com/'

for item in exchangeList :
    data = {
        "title" : item.select_one(".h_lst").get_text(),
        "exchange" : item.select_one(".value").get_text(),
        "change" : item.select_one(".change").get_text(),
        "updown" : item.select_one(".head_info.point_dn > .blind").get_text(),
        "link" : baseUrl + item.select_one("a").get("href")
    }
    exchange_datas.append(data)

exchange_datas
```

```python
import pandas as pd

df = pd.DataFrame(exchange_datas)

df.set_index("title", inplace=True)

df
```

```python
df.to_csv('../data/data03/03. naver_finance.xlsx', sep=',', encoding='utf-8')
```

```python
import urllib
from urllib.request import urlopen, Request
from bs4 import BeautifulSoup as bs

html = "https://ko.wikipedia.org/wiki/{search_words}"

# urllib.parse.quote -> 글자를 URL로 인코딩하는 기능
req = Request(html.format(search_words=urllib.parse.quote("여명의_눈동자")))
res = urlopen(req)

res.status
```

```python
soup = bs(res, "html.parser")

print(soup.prettify())
```

```python
soup.find_all("ul")
```

```python
n = 0
find_char = '채시라'

for each in soup.find_all("ul") :
    print("=>" + str(n) + "=" * 50)
    print(each.get_text())
    if find_char in each.get_text() :
        print(n)
        break
    n += 1
```
